### Tuple
- Ordered
- Immutable
- Index based
- Stores any datatype
- Syntax: ()

````python
# Creation
educators_tup = ('Toba', 'Erioluwa', 'Micheal', 'Ajiboye', 'Iyanu')
students_age_tup = tuple((23, 29, 19, 25, 43, 32, 21, 0.5, 25, 30, 25, 25))


duality_cause = tuple(("Control", 25, True, 0.9))
# print(type(duality_cause), type(educators_tup))


# Common Actions
comb_tuple = educators_tup + duality_cause #concatenating two tuples


# Inbuilt Methods
students_age_tup.count(25)
students_age_tup.index(25)
````

### List
- Ordered
- Mutable
- Stores any datatype
- Index based
- Syntax: []

````python
# Creation
educators_list = ['Toba', 'Erioluwa', 'Micheal', 'Ajiboye', 'Iyanu']
students_age = [23, 29, 19, 25, 43, 32, 21]
num_list = list((1,3,4,5,6,7,3,8,0,2,1,4))
mixed_datatypes = list((True, 'Pharaoh', 'Trump', 23, 94.23))

# Common Actions
print(educators_list[0]) # getting an item using its index
print(educators_list[0:3]) #slicing a list
educators_list[0] = 'Obaloluwa' #reassigning an item in a list with another value.

# Inbuilt Methods
educators_list.reverse()
educators_list.index('Micheal')

## Reversing a list shortcut
# educators_list[::-1]
````

### Dictionaries
- Key-value pairs
- Keys unique and case sensitive
- Keys are immutable and duplicates not allowed.
- Values can be of any data type and can be duplicated.
- Syntax: {key: value}


````python
# Creation
student_A = {"name":"Ebuka","age":50}
student_B = {"name": "Ajiboye", "age":35}
part_time = dict(time = 'night', value= 9.00)


# # Common Actions
student_A["grade"] = "A" # adding a new key value pair to an existing dictionary
student_A["age"] = 30 # changing the value of an existing key in a dictionary

#creating a dict from two lists
new_dict = dict(zip(educators_list, students_age))


# Inbuilt Methods
student_B.values()
student_A.keys()
````

## Control Structures with Data Structures
+ Iterate
+ Filter
+ Transform
+ Apply logic


### Iteration

````python
for x in students_age:
    print(x)
    
````

````python
students_age
````

### Filter

````python
for x in students_age_tup:
    if x < 18:
        print(x)
````

### Transform

````python
print(num_list)
transformed_list = []
for x in num_list:
    x += 20
    transformed_list.append(x)

transformed_list
````

### Apply Logic

````python
even_odd = []
for i in transformed_list:
    if i % 2 == 0:
        print('even')
        even_odd.append('even')
    else:
        print('odd')
        even_odd.append('odd')

even_odd
````
